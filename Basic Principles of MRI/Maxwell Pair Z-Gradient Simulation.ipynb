```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
print("Import completed")
```

```python
# Permeability of free space
mu0 = 4 * np.pi * 1e-7     
```

```python
"""
Coil Parameters
"""
#Current
I = 10  
#Coil radius               
r = 0.35  
#Distance between coils
b = np.sqrt(3) * r      
```

```python
#Generate z-axis points
z = np.linspace(-1.0, 1.0, 1000)
```

```python
#Closed-form magnetic field equation for a Maxwell pair
Bz = (mu0 * I * r**2 / 2) * (
    1 / (r**2 + (z - b/2)**2)**(3/2) - 1 / (r**2 + (z + b/2)**2)**(3/2))
```

```python
#Plot Magnetic Field
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(z, Bz, 'b', linewidth=1.5)

ax.set_xlabel("Z-axis", fontsize=12)
ax.set_ylabel(r"$B_z$ (T)", fontsize=12)
plt.title("Z Magnetic field ")
ax.set_xlim(-0.7, 0.7)

formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-3, -3))
ax.yaxis.set_major_formatter(formatter)

ax.tick_params(direction='in', top=True, right=True)

plt.tight_layout()
plt.show()
```

```python
#Calculate gradient dBz/dz
gradient = np.gradient(Bz, z)
```

```python
#Plot Gradient by maxwell pair
plt.figure(figsize=(8,6))
plt.plot(z, gradient, linewidth=2, color="red", label="Gradient dBz/dz")
plt.xlabel("z (m)")
plt.ylabel("Gradient (T/m)")
plt.title("Z-Gradient Produced by Maxwell Pair")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
