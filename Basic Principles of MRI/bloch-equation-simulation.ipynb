```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 8, 1000)     # 8 seconds
```

```python
M0 = 40
```

```python
Mx0 = M0
My0 = 0
Mz0 = 0
```

```python
def bloch_matrix_solution(T1, T2, t, M0):

    # Relaxation matrices
    E1 = np.exp(-t / T1)
    E2 = np.exp(-t / T2)

    # Magnetization components
    Mx = M0 * E2
    My = np.zeros_like(t)
    Mz = M0 * (1 - E1)

    # Magnitude of transverse magnetization
    Mxy = np.sqrt(Mx**2 + My**2)

    return Mx, My, Mz, Mxy
```

```python
T1_A = 3.0
T2_A = 0.3

MxA, MyA, MzA, MxyA = bloch_matrix_solution(T1_A, T2_A, t, M0)
```

```python
T1_B = 1.5
T2_B = 0.1

MxB, MyB, MzB, MxyB = bloch_matrix_solution(T1_B, T2_B, t, M0)
```

```python
plt.figure(figsize=(8,4))

plt.plot(t*1000, MxyA, 'r', linewidth=2, label='Moiety A')

plt.plot(t*1000, MxyB, 'b', linewidth=2, label='Moiety B')

plt.xlabel('Time (ms)')
plt.ylabel('Transverse Magnetization (A/m)')

plt.xlim(0,5000)
plt.ylim(0,40)

plt.title('Transverse Relaxation')

plt.legend()

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(t*1000, MzA, 'r', linewidth=2, label='Moiety A')

plt.plot(t*1000, MzB, 'b', linewidth=2, label='Moiety B')

plt.xlabel('Time (ms)')
plt.ylabel('Longitudinal Magnetization (A/m)')

plt.xlim(0,5000)
plt.ylim(0,40)

plt.title('Longitudinal Relaxation')

plt.legend()

plt.grid(True)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(111, projection='3d')

ax.plot(MxA, MyA, MzA,
        color='red',
        linewidth=2,
        label='Moiety A')

ax.plot(MxB, MyB, MzB,
        color='blue',
        linewidth=2,
        label='Moiety B')

ax.set_xlabel("Mx")

ax.set_ylabel("My")

ax.set_zlabel("Mz")

ax.legend()

plt.show()
```
