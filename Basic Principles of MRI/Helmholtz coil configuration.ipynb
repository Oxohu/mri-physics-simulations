```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Constants"""

mu0 = 4 * np.pi * 1e-7

I = 100

a = 0.35

d = a

z = np.linspace(-0.4, 0.4, 400)
```

```python
"""Magnetic field from the upper coil"""

B1 = (mu0 * I * a**2) / (2 * (a**2 + (z - d/2)**2)**1.5)
```

```python
"""Magnetic field from the lower coil"""

B2 = (mu0 * I * a**2) / (2 * (a**2 + (z + d/2)**2)**1.5)
```

```python
"""Total magnetic field"""

Bz = B1 + B2
```

```python
"""Plot the magnetic field"""

plt.figure(figsize=(6,4))

plt.plot(z, Bz, color='blue', linewidth=1.5)

plt.xlim(-0.4, 0.4)
plt.ylim(0.5e-4, 3.0e-4)

plt.xlabel("Z-axis (m)")
plt.ylabel("Bz (T)")    
plt.title("Z-component Magnetic Field")

plt.grid(False)

plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.show()
```

```python
"""3D Helmholtz coil configuration"""

from mpl_toolkits.mplot3d import Axes3D
```

```python
theta = np.linspace(0, 2*np.pi, 300)

x = a * np.cos(theta)
y = a * np.sin(theta)

z1 = np.ones_like(theta) * d/2
z2 = -np.ones_like(theta) * d/2
```

```python
fig = plt.figure(figsize=(7,5))

ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z1, color='blue', linewidth=2)
ax.plot(x, y, z2, color='blue', linewidth=2)

ax.plot([0,0],[0,0],[-d/2,d/2],color='blue',linewidth=2)

ax.text(0,0,0,"2s")

ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.4,0.4)
ax.set_zlim(-0.2,0.2)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.view_init(elev=15, azim=-125)

plt.show()
```

```python
"""Print the maximum magnetic field"""

print("Maximum Bz =", np.max(Bz), "Tesla")
```
